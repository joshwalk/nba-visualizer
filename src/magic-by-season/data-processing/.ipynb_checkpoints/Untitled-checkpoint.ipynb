```python
import pandas as pd
df = pd.read_csv("2020-magic-season.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df_filtered = df[['PTS', 'AST', 'TRB', 'BLK', 'STL']]
```

```python
df_pct = df_filtered/df_filtered[df_filtered.columns].sum()*100
df_pct = df_pct.round(decimals=1)
```

```python
df_pct.insert(0, "Name", df["Name"])
```

```python
df_pct
```

```python
df_pct.set_index('Name')
```

```python
df_pct.to_json(path_or_buf="magic-by-season-pct.json", orient="columns", force_ascii=False)
```

```python
df_pct.to_json(orient="records", force_ascii=False)
```

```python
df_transposed = df_pct.transpose()
```

```python
df_transposed
```

```python
df_transposed.to_json(path_or_buf="magic-by-season-pct.json",orient="values", force_ascii=False)
```
