```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
all_years = {}
for yr in range(1990, 2021):
    html = urlopen(f"https://www.basketball-reference.com/teams/ORL/{yr}.html")
    soup = BeautifulSoup(html, "lxml")
    soup_table = soup.find(id="all_totals-playoffs_totals").table
    
    dfs = pd.read_html(str(soup_table))
    df = dfs[0]
    df = df.rename({"Unnamed: 1": "Name"}, axis=1)
    df =  df[['Name', 'PTS', 'AST', 'TRB', 'BLK', 'STL']]
    
    df_transposed = df.transpose().reset_index()
    headers = df_transposed.iloc[0]
    new_df  = pd.DataFrame(df_transposed.values[1:], columns=headers)
    new_df = new_df.rename({"Name": "category"}, axis=1)
    new_df.columns = new_df.columns.fillna('total')
    
    all_years[yr] = list(new_df.to_dict(orient="index").values())
```

```python
import json
f = open("all-years.json", "w")
json.dump(all_years, f, ensure_ascii=False)
f.close()
```
